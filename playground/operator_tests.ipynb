## Example 1: Basic Single Document Processing

Let's start with a simple example processing one document.

```python
import json
```

```python
with open(
        "/mnt/SSD1/nlp/episodic-memory-benchmark/epbench/data/chapter_book.json",
        "r",
    ) as f:
        chapters = json.load(f)
```

```python
book = []
for chapter, text in chapters.items():
    book.append(text)
```

```python
processor = GSWProcessor(
    model_name="gpt-4o",
    enable_coref=True,
    enable_chunking=True,
    enable_context=True,
    chunk_size=3,
    overlap=1,
    coref_chunk_size=20, 
)

print("Processing book...")
gsw_structures = processor.process_documents(book, output_dir="output")
```
